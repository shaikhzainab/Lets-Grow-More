**4)Image to Pencil Sketch with Python:**

We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.


**Step 1 : Importing Libraries**

```python
from google.colab.patches import cv2_imshow
import cv2
```

**Step 2 : Loading the Image.**

```python
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Lets Grow More Dataset/Pedestrain Bridge.jpg',1)
cv2_imshow(image)
```

**Step 3 : Converting the image into grey scale image.**

```python
from IPython.core.display import Image
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image)
```

**Step 4 : Invert Gray scale**

Now the next step is to invert the new grayscale image

We are using the bitwise_not function which is used to make brighter regions lighter and vice versa so that we can find the edges to create a pencil sketch.


```python
img_invert = cv2.bitwise_not(gray_img)
cv2_imshow(img_invert)
```

**Step 5 : Converting to blur image**

The next step in the process is to blur the image by using the Gaussian Function in OpenCV We have used the gaussian blur technique with 21 x 21 pixel and the default sigma values filter on the image to smoothen our image. By increasing the filter size, we can create thin lines for our sketch and it is used to reduce the noise in the image

```python
blur = cv2.GaussianBlur(img_invert,(21,21),sigmaX=0, sigmaY=0)
cv2_imshow(blur)
```

```python
invert_blur = cv2.bitwise_not(blur)
cv2_imshow(invert_blur)
```

**Step 6 : to invert the blurred image, then we can easily convert the image into a pencil sketch:**

```python
sketch = cv2.divide(gray_img,invert_blur,scale=256.0)
cv2_imshow(sketch)
```
